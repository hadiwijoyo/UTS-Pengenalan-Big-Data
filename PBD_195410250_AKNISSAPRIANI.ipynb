```python
#Import terlebih dahulu library dan dataset yang akan digunakan seperti pandas, numpy, seaborn dan matplotlib. 
#selain itu membaca dataset yang telah saya unduh tadi.

import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline

df=pd.read_csv('revisi-KPI-2019-DESEMBER-BIDANG-PELAYARAN.csv')
df.head()

#Visualisasi data yang dilakukan yaitu melihat grafik jumlah penumpang naik dan jumlah penumpang turun 
#di 10 pelabuhan yang ada dikepulauan seribu dengan menggunakan grafik line/garis
```

```python
sns.set(rc={'figure.figsize':(21.7,5.27)})
ay = sns.lineplot(x='tanggal', y='penumpang_turun', data=df)

#Membuat grafik dari data penumpang yang naik selama bulan desember 2019 
#di pelabuhan-pelabuhan yang ada di kepulauan seribu
```

```python
sns.set(rc={'figure.figsize':(21.7,5.27)})
ay = sns.lineplot(x='pelabuhan', y='penumpang_naik', data=df)

#Membuat grafik dari data penumpang yang turun selama bulan desember 2019 
#di pelabuhan-pelabuhan yang ada di kepulauan seribu
```

```python
sns.set(rc={'figure.figsize':(21.7,5.27)})
ay = sns.lineplot(x='pelabuhan', y='kapal_tiba', data=df)

#Membuat grafik dari data pelabuhan dan kapal tiba selama bulan desember 2019 
#di pelabuhan-pelabuhan yang ada di kepulauan seribu
```
